```python
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
import csv
from scipy import stats
```

# The most consuming/producing countries

```python
df_c=pd.read_csv('data_c.csv')
```

```python
df_c
```

```python
df_c=df_c.set_index("countries")
```

```python
remove=df_c[df_c['2000']=='--'].index.tolist()
for i in remove:
     df_c.drop(labels=i,inplace=True, axis=0)
```

```python
df_c.drop(labels='        Former Serbia and Montenegro',inplace=True, axis=0)
```

```python
df_c=df_c.astype(float)
```

```python
df_c=df_c.sort_values(by=['2021'],ascending=False)
```

```python
df_c
```

```python
con=df_c.index.tolist()
```

```python
year = []
i = 1999
while i<2021:
    i=i+1
    year.append(i)
```

```python
fig,axe = plt.subplots(2,1,figsize = (10, 10))
axe[0].bar(con[1:6], df_c['2021'][1:6], color ='maroon',
        width = 0.4)
axe[0].set_title('The most consuming countries in 2021')
axe[0].set_ylabel('E [Joules]')

for i in range(1,6):
    axe[1].plot(year,df_c.iloc[i].tolist(),label=con[i])
axe[1].set_title("Energy consumption in the most consuming countries from 2000 to 2021")
axe[1].set_ylabel('E [Joules]')
plt.legend()
#plt.savefig('energy_consume.png',dpi=2000)
```

# Total energy consumption in 2021

```python
labels = ["Coal", "Natural gas", "Petroleum and other liquids", "Nuclear"
          ,'Renewable and other']
y = np.array([166.72, 150.35, 187.463, 28.31,72.842])
fig,ax = plt.subplots(1,1,figsize=(7,7))
ax.pie(y,labels =labels,autopct='%.1f%%'
       ,wedgeprops={'linewidth': 2.0, 'edgecolor': 'white'})
ax.set_title('Total energy consumption in 2021')

plt.savefig('piechart.png',dpi=1000)
```
